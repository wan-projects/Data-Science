# 1-Latihan SKLearn Decision Tree


```python
import pandas as pd

# Membaca file iris.csv
iris = pd.read_csv('Iris.csv')
```

```python
# Melihat informasi dataset
iris.info()

# Melihat informasi dataset pada 5 baris pertama
iris.head()
```

```python
# Menghilangkan kolom yang tidak penting
iris.drop('Id',axis=1,inplace=True)
```

```python
# Memisahkan atribut dan label
X = iris[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = iris['Species']

# Membagi dataset menjadi data latih & data uji
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=123)
```

```python
from sklearn.tree import DecisionTreeClassifier

#M Membuat model Decision Tree
tree_model = DecisionTreeClassifier()

# Melatih model dengan menggunakan data latih
tree_model = tree_model.fit(X_train, y_train)
```

```python
# Evaluasi Model
from sklearn.metrics import accuracy_score

y_pred = tree_model.predict(X_test)

acc_secore = round(accuracy_score(y_pred, y_test), 3)

print('Accuracy: ', acc_secore)
```

```python
# Prediksi model dengan tree_model.predict([[SepalLength, SepalWidth, PetalLength, PetalWidth]])
print(tree_model.predict([[6.2, 3.4, 5.4, 2.3]])[0])
```

```python
from textwrap import fill
from sklearn.tree import export_graphviz
export_graphviz(
    tree_model,
    out_file= "iris_tree.dot",
    feature_names= ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"],
    class_names= ["Iris-setosa", "Iris-versicolor", "Iris-virginica"],
    rounded= True,
    filled= True
)
```

# 3-Latihan SKLearn Logistic Regression


```python
import pandas as pd

# Membuat dataset dan mengubahnya menjadi dataframe
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Drop kolom yang tidak diperlukan
data = df.drop(columns=['User ID'])

# Jalankan proses one-hot endcoding dengan pd.get_dummies()
data = pd.get_dummies(data)
data
```

```python
# Pisahkan atribut dan label
predictions = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']

X = data[predictions]
y = data['Purchased']
```

```python
# Lakukan normalisasi terhadap data yang kita miliki
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
scaled_data = pd.DataFrame(scaled_data, columns= X.columns)
scaled_data.head()
```

```python
from sklearn.model_selection import train_test_split

# Bagi data menjadi train dan test untuk setiap atribut dan label
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

# Latih model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Uji akurasi model
model.score(X_test, y_test)
```

# 4-Latihan SKLearn K-means


```python
import pandas as pd

# Ubah file csv menjadi dataframe
df = pd.read_csv('Mall_Customers.csv')

# Tampilkan 3 baris pertama
df.head(3)
```

```python

# Ubah nama kolom
df = df.rename(columns={'Gender': 'gender', 'Age': 'age',
                        'Annual Income (k$)': 'annual_income',
                        'Spending Score (1-100)': 'spending_score'})

# Ubah data kategorik menjadi data numerik
df['gender'].replace(['Female', 'Male'], [0,1], inplace=True)

# Tampilkan data yang sudah di preprocess
df.head(3)
```

```python
from sklearn.cluster import KMeans

# Menghilangkan kolom customer id dan gender
X = df.drop(['CustomerID', 'gender'], axis=1)

# Membuat list yang berisi inertia
clusters = []
for i in range(1,11):
  km = KMeans(n_clusters=i).fit(X)
  clusters.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Membuat plot inertia
fig, ax = plt.subplots(figsize=(8, 4))
sns.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
# Membuat objek KMeans
km5 = KMeans(n_clusters=5).fit(X)

# Menambahkan kolom label pada dataset
X['Labels'] = km5.labels_

# Membuat plot KMeans dengan 5 klaster
plt.figure(figsize=(8, 4))
sns.scatterplot(x=X['annual_income'], y=X['spending_score'], hue=X['Labels'],
                palette=sns.color_palette('hls', 5))

plt.title('KMeans dengan 5 Cluster')
plt.show()
```

# 5-Latihan SKLearn PCA


```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()
atribut = iris.data
label = iris.target

# Bagi dataset menjadi train set dan test set
X_train, X_test, y_train, y_test = train_test_split(
    atribut, label, test_size=0.2, random_state=1
)
```

```python
# Model Decision Tree dan menghitung berapa akurasinya tanpa
# Menggunakan PCA
from sklearn import tree

decision_tree = tree.DecisionTreeClassifier()
model_pertama = decision_tree.fit(X_train, y_train)
model_pertama.score(X_test, y_test)
```

```python
# Menggunakan PCA dan menghitung variance dari setiap atribut
from sklearn.decomposition import PCA

# Membuat objek PCA dengan 4 principal component
pca = PCA(n_components=4)

# Mengaplikasikan PCA pada dataset
pca_attributes = pca.fit_transform(X_train)

# Melihat variance dari setiap atribut
pca.explained_variance_ratio_
```

```python
# PCA dengan 2 principal component
pca = PCA(n_components = 2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)
```

```python
# Uji akurasi classifier
model2 = decision_tree.fit(X_train_pca, y_train)
model2.score(X_test_pca, y_test)
```
